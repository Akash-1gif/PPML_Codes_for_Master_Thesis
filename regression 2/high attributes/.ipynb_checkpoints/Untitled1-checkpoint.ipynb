```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
upper_threshold = df['Population'].quantile(0.90)
lower_threshold = df['Population'].quantile(0.10)

# Filter the DataFrame to remove the top 5% and bottom 5% rows based on the Population column
df_filtered = df[(df['Population'] <= upper_threshold) & (df['Population'] >= lower_threshold)]

# Display the filtered DataFrame
df_filtered.shape
```

```python
df_filtered['Population'].max()
```

```python
df_filtered['Population'].min()
```

```python
df_filtered['Population'].median()
```

```python
df_filtered['Population'].mean()
```

```python
df_filtered['Population']=df_filtered['Population']/100000
```

```python
df_filtered['Population'].max()
```

```python
df_filtered['Population'].min()
```

```python
df_filtered.rename(columns={'Population': 'Population (in 100K)'}, inplace=True)
```

```python
df_filtered
```

```python
df_filtered['Population (in 100K)'] = df_filtered['Population (in 100K)'].round(2)
```

```python
df_filtered.head()
```

```python
df=df_filtered
```

```python
df.shape
```

```python
v=df['Country'].value_counts()
```

```python
print(v)
```

```python
for i in v:
    print(i)
```

```python
# Get the value counts of the Country column
country_counts = df['Country'].value_counts()

# Print each country and its count
for country, count in country_counts.items():
    print(f'{country}: {count}')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Country'] = label_encoder.fit_transform(df['Country'])
df.head()
```

```python
label_encoder = LabelEncoder()
df['Status'] = label_encoder.fit_transform(df['Status'])
df.head()
```

```python
df.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
column_names = df.columns
print(column_names)
```

```python
y=df['Life expectancy ']
y.head()
```

```python
X=df.drop(['Life expectancy '],axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
print(f"Model Coefficients: {model.coef_}")
print(f"Model Intercept: {model.intercept_}")
print(f"Mean Squared Error: {mse}")
```

```python
df.to_csv('cleaned_dataframe.csv',index=False)
```
