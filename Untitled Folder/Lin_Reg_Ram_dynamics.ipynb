```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### initial RAM values:

```python
low_attr_varying_inst_ram=[2.66,2.86,3.20]
mid_attr_varying_inst_ram=[2.75,2.90,3.20]
high_attr_varying_inst_ram=[2.80,3.00,3.25]

low_inst_varying_attr_ram=[2.66,2.75,2.80]
mid_inst_varying_attr_ram=[2.86,2.90,3.00]
high_inst_varying_attr_ram=[3.20,3.20,3.25]
```

```python
x_values=[360,680,1080]
plt.xticks(x_values)

plt.plot(x_values,low_attr_varying_inst_ram, marker='o', label='low attr')
plt.plot(x_values,mid_attr_varying_inst_ram, marker='o', label='mid attr')
plt.plot(x_values,high_attr_varying_inst_ram, marker='o', label='high attr')
# Add labels and title
plt.xlabel('Increase in number of instances')
plt.ylabel('ram% util')
plt.title('Line Graph of ram% utilization wrt increase in number of instances')
plt.legend()
plt.show()
```
