```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### CKKS (With rescaling, training time graphs):

```python
low_attr_varying_instance_training_time=[74,138,212]
mid_attr_varying_instance_training_time=[83,155,281]
high_attr_varying_instance_training_time=[87,167,262]

low_inst_varying_attr_training_time=[74,83,87]
mid_inst_varying_attr_training_time=[138,155,167]
high_inst_varying_attr_training_time=[212,281,262]
```

```python
x_values=[360,680,1080]
plt.xticks(x_values)

plt.plot(x_values,low_attr_varying_instance_training_time, marker='o', label='low attr')
plt.plot(x_values,mid_attr_varying_instance_training_time, marker='o', label='mid attr')
plt.plot(x_values,high_attr_varying_instance_training_time, marker='o', label='high attr')
# Add labels and title
plt.xlabel('Increase in number of instances')
plt.ylabel('training time')
plt.title('Line Graph of training time wrt increase in number of instances')
plt.legend()
plt.show()
```

```python
x_values=[8,15,22]
plt.xticks(x_values)

plt.plot(x_values,low_inst_varying_attr_training_time, marker='o', label='low inst')
plt.plot(x_values,mid_inst_varying_attr_training_time, marker='o', label='mid inst')
plt.plot(x_values,high_inst_varying_attr_training_time, marker='o', label='high inst')
# Add labels and title
plt.xlabel('Increase in number of attributes')
plt.ylabel('training time')
plt.title('Line Graph of training time wrt increase in number of attributes')
plt.legend()
plt.show()
```
