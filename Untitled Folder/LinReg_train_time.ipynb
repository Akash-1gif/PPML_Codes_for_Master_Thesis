```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### CKKS (With rescaling, training time graphs):

```python
low_attr_varying_instance_training_time=[3.41,3.21,3.02]
mid_attr_varying_instance_training_time=[13.45,9.12,12.839]
high_attr_varying_instance_training_time=[40.03,27.54,34.76]

low_inst_varying_attr_training_time=[3.41,13.45,40.03]
mid_inst_varying_attr_training_time=[3.21,9.12,27.54]
high_inst_varying_attr_training_time=[3.02,12.839,34.76]
```

```python
x_values=[352,680,1114]
plt.xticks(x_values)

plt.plot(x_values,low_attr_varying_instance_training_time, marker='o', label='datasets with low attr')
plt.plot(x_values,mid_attr_varying_instance_training_time, marker='o', label='datasets with mid attr')
plt.plot(x_values,high_attr_varying_instance_training_time, marker='o', label='datasets with high attr')
# Add labels and title
plt.xlabel('Increase in number of instances')
plt.ylabel('training time')
plt.title('Line Graph of training time wrt increase in number of instances')
plt.legend()
plt.show()
```

```python
x_values=[5,14,21]
plt.xticks(x_values)

plt.plot(x_values,low_inst_varying_attr_training_time, marker='o', label='datasets with low inst')
plt.plot(x_values,mid_inst_varying_attr_training_time, marker='o', label='datasets with mid inst')
plt.plot(x_values,high_inst_varying_attr_training_time, marker='o', label='datasets with high inst')
# Add labels and title
plt.xlabel('Increase in number of attributes')
plt.ylabel('training time')
plt.title('Line Graph of training time wrt increase in number of attributes')
plt.legend()
plt.show()
```
