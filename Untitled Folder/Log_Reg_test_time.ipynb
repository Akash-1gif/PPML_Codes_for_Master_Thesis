```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### CKKS (With rescaling, training time graphs):

```python
low_attr_varying_instance_testing_time=[2.02,3.51,5.16]
mid_attr_varying_instance_testing_time=[2.14,4.17,6.92]
high_attr_varying_instance_testing_time=[2.54,4.09,6.16]

low_inst_varying_attr_testing_time=[2.02,2.14,2.54]
mid_inst_varying_attr_testing_time=[3.51,4.17,4.09]
high_inst_varying_attr_testing_time=[5.16,6.92,6.16]
```

```python
x_values=[90,170,270]
plt.xticks(x_values)

plt.plot(x_values,low_attr_varying_instance_testing_time, marker='o', label='low attr')
plt.plot(x_values,mid_attr_varying_instance_testing_time, marker='o', label='mid attr')
plt.plot(x_values,high_attr_varying_instance_testing_time, marker='o', label='high attr')
# Add labels and title
plt.xlabel('Increase in number of instances')
plt.ylabel('testing time')
plt.title('Line Graph of testing time wrt increase in number of instances')
plt.legend()
plt.show()
```

```python
x_values=[8,15,22]
plt.xticks(x_values)

plt.plot(x_values,low_inst_varying_attr_testing_time, marker='o', label='low inst')
plt.plot(x_values,mid_inst_varying_attr_testing_time, marker='o', label='mid inst')
plt.plot(x_values,high_inst_varying_attr_testing_time, marker='o', label='high inst')
# Add labels and title
plt.xlabel('Increase in number of attributes')
plt.ylabel('testing time')
plt.title('Line Graph of testing time wrt increase in number of attributes')
plt.legend()
plt.show()
```
