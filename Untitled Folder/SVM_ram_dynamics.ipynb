```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### CKKS:

```python
low_attr_varying_inst_ram=[5.18,8.86,11.84]
mid_attr_varying_inst_ram=[4.40,7.77,11.95]
high_attr_varying_inst_ram=[4.84,7.49,11.43]

low_inst_varying_attr_ram=[5.18,4.40,4.84]
mid_inst_varying_attr_ram=[8.86,7.77,7.49]
high_inst_varying_attr_ram=[11.84,11.95,11.43]
```

```python
x_values=[360,680,1080]
plt.xticks(x_values)

plt.plot(x_values,low_attr_varying_inst_ram, marker='o', label='low attr')
plt.plot(x_values,mid_attr_varying_inst_ram, marker='o', label='mid attr')
plt.plot(x_values,high_attr_varying_inst_ram, marker='o', label='high attr')
# Add labels and title
plt.xlabel('Increase in number of instances')
plt.ylabel('ram% util')
plt.title('Line Graph of ram% utilization wrt increase in number of instances')
plt.legend()
plt.show()
```

```python
x_values=[8,15,22]
plt.xticks(x_values)

plt.plot(x_values,low_inst_varying_attr_ram, marker='o', label='low inst')
plt.plot(x_values,mid_inst_varying_attr_ram, marker='o', label='mid inst')
plt.plot(x_values,high_inst_varying_attr_ram, marker='o', label='high inst')
# Add labels and title
plt.xlabel('Increase in number of attributes')
plt.ylabel('ram% util')
plt.title('Line Graph of ram% utilization wrt increase in number of attributes')
plt.legend()
plt.show()
```
