```python
top_countries = df['Country'].value_counts().index[:30]

# Step 2: Filter the DataFrame to include only rows where 'Countries' is in the top 40
filtered_df = df[df['Country'].isin(top_countries)]
```

```python
X,y=filtered_df.drop(['Life expectancy '],axis=1),filtered_df['Life expectancy ']
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=tts(X,y, test_size=0.2, random_state=69)
X_train.to_csv('X_train1.csv', index=False)
X_test.to_csv('X_test1.csv', index=False)
y_train.to_csv('y_train1.csv', index=False)
y_test.to_csv('y_test1.csv', index=False)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
```
