```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('Employee.csv')
df.head()
```

```python
label_encoder = LabelEncoder()

df['Edu'] = label_encoder.fit_transform(df['Education'])
```

```python
label_encoder = LabelEncoder()

df['city'] = label_encoder.fit_transform(df['City'])
```

```python
label_encoder = LabelEncoder()

df['gender'] = label_encoder.fit_transform(df['Gender'])
```

```python
label_encoder = LabelEncoder()

df['benched'] = label_encoder.fit_transform(df['EverBenched'])
```

```python
df.head()
```

```python
columns_to_drop = ['Education', 'City', 'Age', 'Gender', 'EverBenched']
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
df.shape
```

```python
y=df['LeaveOrNot']
X=df.drop(['LeaveOrNot'],axis=1)
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=800)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier(max_iter=3, tol=None, random_state=42)
```

```python
sgd_clf.fit(X_train, y_train)
```

```python
y_pred = sgd_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
